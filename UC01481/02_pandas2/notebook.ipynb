```python
def unif_STATUS_CONF(value):
    if value == "benign":
        return 1
    elif value == "not_enough_data":
        return 2
    elif value == "validated":
        return 3
    else:
        return 4
```

```python
df["STATUS_CONF"] = df["STATUS_CONF"].apply(unif_STATUS_CONF)
```
